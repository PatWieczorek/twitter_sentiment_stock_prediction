```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
amzn = pd.read_csv(f'../datasets/v2.1/more_cols_with_stock/no_weight/amzn.csv')
tsla = pd.read_csv(f'../datasets/v2.1/more_cols_with_stock/no_weight/tsla.csv')
msft = pd.read_csv(f'../datasets/v2.1/more_cols_with_stock/no_weight/msft.csv')
aapl = pd.read_csv(f'../datasets/v2.1/more_cols_with_stock/no_weight/aapl.csv')
goog = pd.read_csv(f'../datasets/v2.1/more_cols_with_stock/no_weight/goog.csv')
googl = pd.read_csv(f'../datasets/v2.1/more_cols_with_stock/no_weight/googl.csv')
```

```python
amzn[['open', 'high', 'low', 'close']] = amzn[['open', 'high', 'low', 'close']]/20
amzn["diffs"] = amzn.close.pct_change() * 100

amzn = amzn.dropna()

bin_2 = pd.cut(amzn.diffs, [-1000, 0, 1000], labels=[-1, 1])
amzn["bin_2"] = bin_2.copy()

bin_3 = pd.cut(amzn.diffs, [-1000, -0.13, 0.15, 1000], labels=[-1, 0, 1])
amzn["bin_3"] = bin_3.copy()

sns.countplot(x=bin_3)
plt.show()
```
